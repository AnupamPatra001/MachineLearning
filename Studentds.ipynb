```python
# outlier detection by indexes
# we know that 
# UF=Q3+1.5*IQR
# LF=Q1-1.5*IQR
```

```python
dfcpy=df.copy().drop(['G1','G2','G3'], axis=1)
dfcpy
```

```python
def detect_outlier(columns):
    outlier_indices=[]
    for column in columns:
        Q1=dfcpy[column].quantile(0.25)
        Q3=dfcpy[column].quantile(0.75)
        IQR=Q3-Q1

        # mask is a boolean condition
        Mask= (dfcpy[column] >= (Q1-1.5*IQR)) & (dfcpy[column] < (Q3+1.5*IQR))
        Mask= Mask.to_numpy() #boolean value to numpy array

        # Find the indices that do not satisfy the mask is the outlier
        false_indices=np.argwhere(~Mask)
        outlier_indices.append(false_indices)
    return np.unique(np.concatenate(outlier_indices).ravel())

# np.concatenate(outlier_indices) >> it combines all the outlier index arrays to single array
# so each column have each array
# .ravel() >> flattens the multi-dimensional array intp 1D array
```
