```python
# All the categorical data are nominal data so we will be using one hot encoding
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder()
```

```python
encoder
```

```python
# encoding
encoded_data=encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray()
```

```python
encoder.get_feature_names_out()
```

```python
encodedDf=pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out())
encodedDf
```

```python
# As we have the encoded data for categorical column,
# we can drop the categorical data to clean the data
df.drop(['Airline', 'Source', 'Destination', 'Additional_Info'], axis=1, inplace= True)
```
